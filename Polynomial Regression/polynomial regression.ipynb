## 1.Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2.Load the Dataset

```python
dataset=pd.read_csv("position_salaries.csv")
dataset.head()
```

```python
dataset.info()
```

The dataset contains position levels and corresponding salaries.

## 3.Feature and Target Selection

```python
X=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
```

X → Position level (independent variable)

y → Salary (target variable)

Used 1:2 to keep X as a 2D array.

X is a 2D array because machine learning models expect input features in matrix form, even when there is only one feature.

## 4. Train–Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=0)
```

## 5.Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

Linear Regression is used as a baseline model to compare with Polynomial Regression.

## 6. Visualizing Linear Regression Results

```python
def viz_linear():
    plt.scatter(X,y,color="red")
    plt.plot(X,lin_reg.predict(X),color="blue")
    plt.title("Linear Regression model")
    plt.xlabel("position level")
    plt.ylabel("salary")
    plt.show()
    return
```

```python
viz_linear()
```

Linear Regression underfits because salary does not increase linearly with position level.

## 7. Polynomial Feature Transformation and model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
pol_reg=LinearRegression()
pol_reg.fit(X_poly,y)
```

Polynomial features allow the model to capture non-linear relationships.Although it uses Linear Regression internally, polynomial features make the model non-linear.

## 9.Visualizing Polynomial Regression Results

```python
def viz_polynomial():
    plt.scatter(X,y,color="red")
    plt.plot(X,pol_reg.predict(poly_reg.fit_transform(X)),color="blue")
    plt.title("Linear regression withpolynomial with degree 4")
    plt.xlabel("position level")
    plt.ylabel("salary")
    plt.show()
    return
    
```

```python
viz_polynomial()
```

Polynomial Regression fits the data much better than Linear Regression.

## 10. Salary Prediction for a New Position Level

```python
lin_reg.predict([[5.5]])
```

```python
pol_reg.predict(poly_reg.fit_transform([[5.5]]))
```

Polynomial Regression provides a more realistic salary prediction for intermediate position levels.
